```python
import netCDF4 as nc
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import datetime 
%matplotlib inline
```

```python
file = '/home/2017LMSOM2/e21608618/TÃ©lÃ©chargements/ncep_msl_1948to2017.nc'
data = Dataset(file,'r',format='NETCDF4')

lon = np.array(data.variables['longitude'][:])
lat = np.array(data.variables['latitude'][:])
t = np.array(data.variables['time'][:])
p = np.array(data.variables['msl'][:])/100
pp = p-np.mean(p, axis=0)
pp[pp>-40]=np.nan

# ul = np.array(data.variables['ulwrf'][:])
# dl = np.array(data.variables['dlwrf'][:])
# ds = np.array(data.variables['dswrf'][:])
# lh = np.array(data.variables['lhtfl'][:])
# sh = np.array(data.variables['shtfl'][:])
```

```python
data.variables['time']
dateref = datetime.datetime.strptime('1948-1-1',"%Y-%m-%d")
time = []
for i in range(0,len(t)):
    time.append(dateref + datetime.timedelta(hours = int(t[i])))
time = np.array(time)

LON, LAT = np.meshgrid(lon,lat)
```

```python


datedebut = datetime.datetime.strptime('1999-12-25',"%Y-%m-%d")
datefin = datetime.datetime.strptime('1999-12-28',"%Y-%m-%d")
ind = np.where((time>=datedebut)&(time<=datefin))[0]

for k in ind:
     
    plt.figure()
    plt.pcolormesh(LON,LAT,pp[k])
    plt.colorbar()
    plt.clim([-60,-40])
    plt.title(time[k])
      
#     plt.figure();

# #     n, bins, patches = plt.hist(p[k].ravel(), bins=50);
#     plt.xlim(-60)
#     plt.show()

```
