```python
import netCDF4 as nc
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
file = '/home/2017LMSOM2/e21608618/Téléchargements/ncep_flux_1979to2017_2017-10-01.nc'
data = Dataset(file,'r',format='NETCDF4')
```

```python
import datetime 
datetime.datetime.fromtimestamp(28000)
```

```python
lon = np.array(data.variables['longitude'][:])
lat = np.array(data.variables['latitude'][:])
t = np.array(data.variables['time'][:])
us = np.array(data.variables['uswrf'][:])
ul = np.array(data.variables['ulwrf'][:])
dl = np.array(data.variables['dlwrf'][:])
ds = np.array(data.variables['dswrf'][:])
lh = np.array(data.variables['lhtfl'][:])
sh = np.array(data.variables['shtfl'][:])

heat = lh +sh
heat = heat - np.mean(heat, axis = 0)
```

```python
data.variables['time']
dateref = datetime.date(1979, 1, 1)
time = []
for i in range(0,len(t)):
    time.append(dateref + datetime.timedelta(int(t[i])))
```

```python
LON, LAT = np.meshgrid(lon,lat)

datedebut = datetime.date(2010, 2, 26)
datefin = datetime.date(2010, 3, 1)

for k in range((datedebut-dateref).days, (datefin-dateref).days+1):

    plt.figure()
    
    plt.pcolormesh(LON,LAT,heat[k])
    
    plt.colorbar()
    plt.title(time[k])
    
    plt.show()

```
