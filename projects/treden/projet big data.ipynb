```python
import netCDF4 as nc
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import datetime 
%matplotlib inline
```

```python
file = 'ncep_msl_1948to2017.nc'
data = Dataset(file,'r',format='NETCDF4')
```

```python
lon = np.array(data.variables['longitude'][:])
lat = np.array(data.variables['latitude'][:])
t = np.array(data.variables['time'][:])
p = np.array(data.variables['msl'][:])/100
# ul = np.array(data.variables['ulwrf'][:])
# dl = np.array(data.variables['dlwrf'][:])
# ds = np.array(data.variables['dswrf'][:])
# lh = np.array(data.variables['lhtfl'][:])
# sh = np.array(data.variables['shtfl'][:])
```

```python
data.variables['time']
dateref = datetime.datetime.strptime('1948-1-1',"%Y-%m-%d")
time = []
for i in range(0,len(t)):
    time.append(dateref + datetime.timedelta(hours = int(t[i])))
time = np.array(time)
```

```python
LON, LAT = np.meshgrid(lon,lat)

datedebut = datetime.datetime.strptime('1998-12-26',"%Y-%m-%d")
datefin = datetime.datetime.strptime('1998-12-28',"%Y-%m-%d")
ind = np.where((time>=datedebut)&(time<=datefin))[0]

for k in ind:

    plt.figure()
    
    plt.pcolormesh(LON,LAT,p[k]-np.mean(p, axis=0))
    
    plt.colorbar()
    plt.title(time[k])
    
    plt.show()

```
